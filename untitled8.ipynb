```python
# Load modules and data
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt

# 1) Dados e modelo
data = sm.datasets.stackloss.load()
data.exog = sm.add_constant(data.exog)

rlm_model = sm.RLM(data.endog, data.exog, M=sm.robust.norms.HuberT())
rlm_results = rlm_model.fit()

# 2) Extrair tabelas do resumo de forma robusta
# - summary2() gera um objeto com HTML consistente
s2 = rlm_results.summary2()
html = s2.as_html()

# LÃª todas as tabelas do HTML para uma lista de DataFrames
tables = pd.read_html(html, header=0)

# Tipicamente:
# tables[0] -> InformaÃ§Ãµes do modelo (Dep. Variable, No. Observations, etc.)
# tables[1] -> Tabela de coeficientes (Coef., Std.Err., z, P>|z|, [0.025, 0.975])
# (a ordem pode variar conforme a versÃ£o)

# 3) Padronizar nomes das colunas da tabela de coeficientes (se existir)
coef_df = None
for t in tables:
    cols = [str(c).strip().lower() for c in t.columns]
    if ("coef." in cols or "coef" in cols) and ("p>|z|" in cols or "p>|t|" in cols):
        coef_df = t.copy()
        coef_df.columns = cols  # normaliza cabeÃ§alhos para minÃºsculas
        break

# 4) VisualizaÃ§Ã£o dos coeficientes com IC 95%
if coef_df is not None:
    # Tentar detectar nomes de colunas presentes em versÃµes diferentes
    # coef, std.err., z/t, p>|z|/p>|t|, [0.025, 0.975]
    def pick(*cands):
        for c in cands:
            if c in coef_df.columns:
                return c
        return None

    c_coef = pick("coef.", "coef")
    c_se   = pick("std.err.", "std err", "std_err")
    c_low  = pick("[0.025", "[0.025, 0.975")  # em algumas versÃµes a coluna vem âcoladaâ
    c_high = pick("0.975]", "0.975]")

    # Se o IC veio numa Ãºnica coluna â[0.025, 0.975]â, quebramos em duas
    if "[0.025, 0.975]" in coef_df.columns:
        split_ci = coef_df["[0.025, 0.975]"].astype(str).str.replace("[","").str.replace("]","")
        coef_df["[0.025"] = split_ci.str.split(",", expand=True)[0].astype(float)
        coef_df["0.975]"] = split_ci.str.split(",", expand=True)[1].astype(float)
        c_low, c_high = "[0.025", "0.975]"

    # Converte numÃ©ricos
    for col in [c_coef, c_se, c_low, c_high]:
        if col and col in coef_df.columns:
            coef_df[col] = pd.to_numeric(coef_df[col], errors="coerce")

    # Ãndice/categorias (primeira coluna costuma ser o nome do termo)
    term_col = coef_df.columns[0]

    # Plot (coeficiente com barras de erro do IC)
    plt.figure(figsize=(8, 5))
    x = range(len(coef_df))
    y = coef_df[c_coef]

    # Se tivermos ICs, usamos como barras de erro
    if c_low and c_high:
        yerr_lower = y - coef_df[c_low]
        yerr_upper = coef_df[c_high] - y
        import numpy as np
        yerr = np.vstack([yerr_lower, yerr_upper])
        plt.errorbar(x, y, yerr=yerr, fmt='o', capsize=4)
    else:
        plt.plot(x, y, 'o')

    plt.xticks(x, coef_df[term_col], rotation=45, ha='right')
    plt.ylabel("Coeficiente (RLM)")
    plt.title("Coeficientes e IC 95% â RLM (summary2)")
    plt.tight_layout()
    plt.show()

    # Exibir as tabelas principais
    print("Tabela de informaÃ§Ãµes do modelo (se encontrada):")
    if len(tables) > 0:
        display(tables[0].head())
    print("\nTabela de coeficientes:")
    display(coef_df)
else:
    print("NÃ£o encontrei uma tabela de coeficientes no summary2(). Verifique a versÃ£o do statsmodels.")
```
